## Summarizing the important concepts from matplot library in python

### The most important module of matplotlib is pyplot. This module provides all the required tools to construct and represent a graph. The construction happens in backend and the representation happens in frontend. For our purpose, the frontend is jupyter notebook itself, so we use a magic %matplotlib inline. The rest of the notebook deals with backend.

```python
# Directly import the most important module from matplotlib
from matplotlib import pyplot as plt
import numpy as np # Because we want number to be plotted
%matplotlib inline
```

```python
# Lets get some data
data1_x = np.array([1,2,3,4,5])
data1_y = np.array([34, 56, 23, 10, 47])
data2_x = np.array([1,2,3,4,5])
data2_y = np.array([67, 25, 39, 18, 20])
```

### The most important thing to realize here is pyplot is state based program. Whatever you write is going to be staged and remebered in the backend. For out purpose, we need to deal with two state variables called figure and axes. The Figure variable is the canvas and axes are the plots inside the canvas. In the simpleset case, where we are plotting only one graph, we dont need to create the figure and axes explicitly, they will be created in the backend by them self. Lets see that now.

```python
plt.plot(data1_x, data1_y, 'g-', data2_x, data2_y, 'r*')
```

```python
plt.show() #This is not requred with the inline magic.
```

### Note the two things above. The state vaiables figure and axes got created in the backend. And only axes in the figure canvas took the default parameters for each variable. Lets change those parameter a little bit. 

```python
plt.title("The figure variable for title for total canvas")
plt.xlabel("The axes varibable for x-axis associated with the only axes in figure")
plt.xlabel("The axes varibable for y-axis associated with the only axes in figure")
```

### What happen above is the cration of figure and axes again, re-writing the previous state. We are having state based approach with pyplot module (remember?)

```python
# Lets now make the graph on the axes
plt.plot(data1_x, data1_y, 'g-', data2_x, data2_y, 'r*')
plt.title("The figure variable for title for total canvas")
plt.xlabel("The axes varibable for x-axis associated with the only axes in figure")
plt.xlabel("The axes varibable for y-axis associated with the only axes in figure")
```

### Now lets move to the important part, creating subplots where we explicitly deal with the hidden state variables figure and axes. In most of the applications, you will need to present multiple graphs on one canvas.

```python
# In case you want many plots without directly having an acess to figure and axes, you can use plt.subplot(nrows, ncols, index)
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize=(6,6)) # Most of the attributes of figure as been assigned here already.
# now for each subplots, we set the plot attributes, the most important ones are data itself, the labels and the title.
ax[0][0].plot(data1_x, data1_y, "bo")
ax[0][0].set_title("Data 1 graph")

ax[0][1].barh(data1_x, data1_y, color = "blue")
ax[0][1].set_title("Analysis for data1")

ax[1][0].plot(data2_x, data2_y, "r--")
ax[1][0].set_title("Data 2 graph")

ax[1][1].barh(data2_x, data2_y, color = "red")
ax[1][1].set_title("Analysis for data2")
```
